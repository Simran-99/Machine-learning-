<h3>Data Preprocessing</h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('breast_cancer.csv')
```

```python
df.head()
```

```python
df.drop(['Sample code number'],axis=1,inplace=True)
```

```python
df.head()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

<h3>Logistic Regression</h3>

```python
from sklearn.linear_model import LogisticRegression 
lr=LogisticRegression(random_state=0)
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cmlr=confusion_matrix(y_test,y_pred)
print(cmlr)
acslr=accuracy_score(y_test,y_pred)
print(acslr)
```

<h3>K-NN Model</h3>

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=100,p=2)
knn.fit(x_train,y_train)
```

```python
y_pred_knn=knn.predict(x_test)
print(np.concatenate((y_pred_knn.reshape(len(y_pred_knn),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm_knn=confusion_matrix(y_test,y_pred_knn)
print(cm_knn)
acs_knn=accuracy_score(y_test,y_pred_knn)
print(acs_knn)
```

<h3>Support vector machine</h3>

```python
from sklearn.svm import SVC
model_svc=SVC()
model_svc.fit(x_train,y_train)
```

```python
y_pred_svc=model_svc.predict(x_test)
```

```python
print(np.concatenate((y_pred_svc.reshape(len(y_pred_svc),1),y_test.reshape(len(y_test),1)),1))
```

```python
cm_svc=confusion_matrix(y_pred_svc,y_test)
print(cm_svc)
acs_svc=accuracy_score(y_pred_svc,y_test)
print(acs_svc)
```
